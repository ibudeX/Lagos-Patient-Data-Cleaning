```python
df1['Gender'] = df1['Gender'].replace({
    'male': 'Male',
    'M': 'Male',
    'female': 'Female',
    'F': 'Female',
    np.nan: 'Undisclosed'
})
```

```python
diagnosis_groups = {
    'Flu': ['flu', 'fyu', 'fou', 'vlu', 'blu', 'elu', 'plu', 'fru'],
    'Anemia': ['anemia', 'akemia', 'anedia', 'anemua', 'anevia', 'adema', 'anecia', 'gnemia', 'ademia'],
    'Cough': ['cough', 'coueh', 'coqgh', 'cosgh', 'cojgh', 'coumh', 'ckugh', 'comgh', 'tough', 'pough', 'coudh', 'coegh', 'jough'],
    'Covid-19': ['covid-19', 'covidg19', 'covcd-19', 'cwvid-19', 'covid19', 'covidg-19', 'covid', 'covcd19'],
    'Hypertension': ['hypertension', 'hypertwnsion', 'pypertension', 'hypertensien', 'hyperteqsion', 'hypertensiln', 'hypertession', 'hypertenslon'],
    'Ulcer': ['ulcer', 'klcer', 'uvcer', 'ugcer', 'ulfer', 'uqcer', 'ulcbr'],
    'Typhoid': ['typhoid', 'tykhoid', 'typhyid', 'typhkid', 'typhord', 'typqoid', 'tyshoid', 'typoid', 'qyphoid', 'typhobd', 'typyoid'],
    'Asthma': ['asthma', 'astma', 'asthxa', 'asthqa', 'asihma', 'asqhma', 'vsthma', 'astmia', 'astxia', 'asthsma', 'astca', 'astsmia', 'astsca', 'asthca', 'astsha', 'asthsqa', 'astsma', 'asthia'],
    'Malaria': ['malaria', 'malqria', 'mafaria', 'malarla', 'kalaria', 'malaira'],
    'Diabetes': ['diabetes', 'dbabetes', 'diabetds', 'diabotes', 'dnabetes', 'diabetbs', 'diabeters', 'ciabetes', 'wiabetes', 'biabetes', 'diobetes', 'eiabetes', 'diacetes', 'dirbetes'],
    'Others': ['others', 'olu']
}
```

```python
diagnosis_mapping = {}
for correct, spellings in diagnosis_groups.items():
    for s in spellings:
        diagnosis_mapping[s.lower()] = correct
    diagnosis_mapping[correct.lower()] = correct
```

```python
def clean_diagnosis_name(diagnosis):
    if pd.isna(diagnosis):
        return 'Undisclosed'
    diagnosis = diagnosis.strip().lower()
    return diagnosis_mapping.get(diagnosis, 'Others')
```

```python
df1['Diagnosis'] = df1['Diagnosis'].apply(clean_diagnosis_name)
```

```python
df1['Diagnosis'].unique()
```

```python
state_groups = {
    'Lagos': ['lagos', 'lagds', 'lagss', 'lagjs', 'aagos', 'eko'],
    'Edo': ['edo'],
    'Ogun': ['ogun', 'ooun', 'oqun', 'opo'],
    'Oyo': ['oyo', 'iyo', 'wyo', 'owo', 'oso'],
    'Kano': ['kano', 'kmo', 'kpno', 'khno', 'kavo', 'kabo', 'kako'],
    'Kaduna': ['kaduna', 'eaduna', 'keduna', 'kadufa'],
    'Imo': ['imo', 'emo', 'iko'],
    'Abuja': ['abuja', 'absja', 'ablja', 'ubhja', 'mbuja', 'ubuja', 'abhja'],
    'Rivers': ['rivers', 'rijers', 'riverss'],
    'Cross River': [
        'cross river', 'cross riser', 'cross rzver', 'crosv river',
        'cross siver', 'crsss river', 'croas river'
    ]
}
```

```python
state_mapping = {}
for correct, spellings in state_groups.items():
    for s in spellings:
        state_mapping[s.lower()] = correct
    state_mapping[correct.lower()] = correct
```

```python
def clean_state_name(state):
    if pd.isna(state):
        return 'Undisclosed'
    state = state.strip().lower()
    return state_mapping.get(state, 'Others')
```

```python
df1['State_of_Residence'] = df1['State_of_Residence'].apply(clean_state_name)
```
